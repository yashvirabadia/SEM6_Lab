<h2>Section 2 : Data Cleaning & Preprocessing</h2>

### 10. Check whether data is null or not

```python
df.isnull().sum()
```

### 11. Convert Age from Days to Years

```python
df['age_years'] = (df['age'] / 365).astype(int)
df[['age', 'age_years']].head(5)
```

### 12. Identifying & Handling Outliers

```python
# Detect invalid blood pressure readingsabs
df[df['ap_hi'] < df['ap_lo']]
```

```python
#Normal human blood pressure:
#Systolic(ap_hi): 90 to 200
#Diastolic(ap_lo): 50 to 140
df = df[df['ap_hi'] >= df['ap_lo']]
df = df[(df['ap_hi'].between(90, 200)) & (df['ap_lo'].between(50, 140))]
df
```

### . Encoding Categorical Variables

```python
df = pd.get_dummies(df, columns=['gender'], drop_first=True)
```

### . Normalization / Scaling of Numerical Features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cols = ['age_years','height','weight','ap_hi','ap_lo']
df[cols] = scaler.fit_transform(df[cols])
```

### . Check Target Distribution

```python
df['cardio'].value_counts()
```

### . Cardio distribution

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
df['cardio'].value_counts().plot(kind='bar')
plt.title("Target Variable Distribution (Cardio)")
plt.xlabel("Cardio (0 = No Disease, 1 = Disease)")
plt.ylabel("Count")
plt.show()
```

### . Age distribution

```python
plt.figure(figsize=(6,4))
plt.hist(df['age_years'], bins=30)
plt.title("Age Distribution")
plt.xlabel("Age (Years)")
plt.ylabel("Frequency")
plt.show()
```

### . Blood Pressure Scatter Plot (ap_hi vs ap_lo)

```python
plt.figure(figsize=(6,4))
plt.scatter(df['ap_hi'], df['ap_lo'], s=10)
plt.title("Systolic vs Diastolic BP")
plt.xlabel("Systolic (ap_hi)")
plt.ylabel("Diastolic (ap_lo)")
plt.show()
```

### . Count Plot for Cholesterol

```python
plt.figure(figsize=(6,4))
df['cholesterol'].value_counts().sort_index().plot(kind='bar')
plt.title("Cholesterol Level Counts")
plt.xlabel("Cholesterol Level (1–3)")
plt.ylabel("Count")
plt.show()
```

### . Height vs Weight Chart

```python
plt.figure(figsize=(6,4))
plt.scatter(df['height'], df['weight'], color='purple')
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,4))

plt.hist(df['height'], label='Height Count', color='blue')
plt.hist(df['weight'], label='Weight Count', color='red')

plt.title("Height & Weight Count Distribution")
plt.xlabel("Value")
plt.ylabel("Count")
plt.legend()
plt.show()
```

### . Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure()
sns.heatmap(df.corr())
plt.title("Heatmap")
plt.show()
```

### . Box Plot for ap_hi

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.boxplot(df['ap_hi'], patch_artist=True,
            boxprops=dict(facecolor='lightblue'),
            medianprops=dict(color='red'))

plt.title("Boxplot of Systolic Blood Pressure (ap_hi)")
plt.ylabel("BP Value")
plt.show()
```
